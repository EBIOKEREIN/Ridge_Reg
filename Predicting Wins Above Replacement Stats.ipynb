```python
predictors = list(required_columns[sfs.get_support()])
```

```python
#CREATING A BACK TEST FUNCTION THAT HELPS CREATE PREDICTIONS
```

```python
def backtest(data, model, predictors, start=5, step = 1):
    all_predictions = []
    years = sorted(data['Season'].unique())
    
    for i in range (start, len(years), step):
        current_year = years[i]
        train = data[data['Season'] < current_year]
        test = data[data['Season'] == current_year]
        train[predictors] = train[predictors].fillna(train[predictors].mean())

        model.fit(train[predictors],train["Next WAR"])
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([preds,test["Next WAR"]], axis=1)
        combined.columns = ["Actual","Predicted"]
        
        all_predictions.append(combined)
        return pd.concat(all_predictions)
```

```python
predictions = backtest(batting, rr, predictors)
```

```python
#Creating new predictors via the player_past function
```

```python
import numpy as np

def players_past(df):
    df = df.sort_values("Season")
    df["player_season"] = range(0,df.shape[0])
    
    df["war_corr"] = list(df[["player_season","WAR"]].expanding().corr().loc[(slice(None),"player_season"),"WAR"])
    df.loc[:, "war_corr"] = df["war_corr"].fillna(1)
    
    if 'war_diff' in df.columns:
        df.loc[df["war_diff"] == np.inf, "war_diff"] = 1
        df["war_diff"] = df["war_diff"].fillna(1)
    else:
        df["war_diff"] = 1  

    df.loc[:, "war_diff"] = df["war_diff"].fillna(1)
    return df

batting = batting.groupby("IDfg", group_keys=False, as_index=False).apply(players_past)
```

```python
def group_averages(df):
    return df["WAR"] / df["WAR"].mean()
```

```python
batting["war_season"] = batting.groupby(by="Season", group_keys=False).apply(group_averages)
```

```python
new_predictors = predictors + ["player_season","war_season","war_diff","war_corr"]
```

```python
fresh_predictions = backtest(batting,rr,new_predictors)
```

```python
mean_absolute_error(fresh_predictions["Actual"],fresh_predictions["Predicted"])
```
